```python
import pandas as pd
import numpy as np

age = pd.read_excel(r'D:\EXIMBANK\Document\XHTDNB\Data\age.xlsx')
df= age.groupby('GOOD/BAD').agg({'AGE':['mean','max','min','median','std','count']}).round(decimals = 2)
print(df)
```

```python
from matplotlib.pyplot import figure
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np


x = age['AGE'].loc[age['GOOD/BAD'].isin([0])]
y = age['AGE'].loc[age['GOOD/BAD'].isin([1])]



bins = 300


figure(figsize=(12,6 ), dpi=80)
pyplot.hist(x, bins, alpha=0.5, label='Good',color = 'mediumblue')
pyplot.hist(y, bins, alpha=0.5, label='Bad',color = 'darkorange')
pyplot.title("Tiêu chí tuổi khách háng")
#pyplot.grid()
pyplot.legend(loc='upper right')
plt.xlabel("Age")
plt.ylabel("Frequenly")
pyplot.axvline(age['AGE'].mean(), color='red', linestyle='dashed', linewidth=1.5)
plt.text(45, 1400, r'<-- Mean',color = 'red')

pyplot.show()
```
